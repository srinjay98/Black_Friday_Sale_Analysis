```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('D:/dataSets/BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar' , figsize = (12,6) , title = 'Purchase Distribution by age')
```

```python
# number of M and F in different age group
sns.set(rc = {'figure.figsize' : (12,6)})
sns.countplot(x = 'Age' , hue = 'Gender' , data = df)
```

```python
sns.set(rc = {'figure.figsize' : (12,6)})
sns.countplot(x = 'Gender' , hue = 'Age' , data = df)
```

```python
sns.set(rc = {'figure.figsize' : (6,6)})
sns.countplot(x = 'Gender' , hue = 'Marital_Status' , data = df)
```

```python
sns.countplot(x = df['City_Category'] , hue = 'Age' , data = df)
```

```python
df.groupby('City_Category').size().plot(kind = 'pie' , autopct = '%0.1f')
```

```python
sns.countplot(x = 'City_Category' , hue = 'Gender' , data = df)
```

```python
df.groupby('City_Category')['Purchase'].size().plot(kind = 'pie' , autopct = '%0.1f')
```

```python
df.groupby('City_Category')['Purchase'].mean().plot(kind = 'pie' , autopct = '%0.1f')
```
